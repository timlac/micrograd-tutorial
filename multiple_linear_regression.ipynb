```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import the 3D plotting module
```

```python
import numpy as np

# Set a random seed for reproducibility
np.random.seed(0)

# Define the number of samples and features
n_samples = 100
n_features = 2  # You can change this to the number of features you want

# Generate random data for multivariate linear regression
# Create random feature values
X = np.random.rand(n_samples, n_features)

# Generate random coefficients for the features
true_coefficients = np.random.rand(n_features)

# Generate random noise for the target variable
noise = 0.1 * np.random.randn(n_samples)

# Calculate the target variable
y = np.dot(X, true_coefficients) + noise

# You can adjust the noise level or other parameters as needed

# Print the true coefficients (for reference)
print("True Coefficients:", true_coefficients)
```

```python
# Create a 3D plot
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot the data in 3D
ax.scatter(X[:, 0], X[:, 1], y, c='b', marker='o')

# Set labels for the three axes
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Target')

# Set a title for the plot
ax.set_title('Generated Data for Multivariate Linear Regression (3D)')

plt.show()
```
